```python
import pandas as pd

data=pd.DataFrame.from_csv("winemag-data_first150k.csv")
```

```python
df=data[['description','country','designation','variety','price','winery','points']]
```

```python
df.head()
```

```python
df.drop_duplicates(subset=None, keep='first', inplace=False)
```

```python
df1=df.dropna()
```

```python
df1
```

```python
def transform_points_simplified(points):
    if points< 84:
        return 1
    elif points >= 84 and points < 88:
        return 2
    elif points >= 88 and points < 92:
        return 3
    elif points >= 92 and points < 96:
        return 4
    else:
        return 5
    
df1=df1.assign(points_simplified = df1['points'].apply(transform_points_simplified))
df1.head()
```

Description Vectorization:

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X=df1['description']
y=df1['points_simplified']
vectorizer=CountVectorizer()
vectorizer.fit(X)
```

```python
X=vectorizer.transform(X)
print("Shape of sparse Matrix is", X.shape)
#print("Amount of non-zero occurrences is", X.nnz)
```

Split the dataset into train and testing set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)

##testing model
predictions=rfc.predict(X_test)
print(classification_report(y_test,predictions))
```

Result with CountVectorizer: 81% precision, 80% recall and f1-score with 0.79 which are not good

Hence, I try different vectorizer(TF-IDF)

```python
###Try TF-IDF
X=df1['description']
y=df1['points_simplified']

vectorizer=TfidfVectorizer()
vectorizer.fit(X)
X=vectorizer.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)

# Testing model
predictions = rfc.predict(X_test)
print(classification_report(y_test, predictions))
```

Result with TF-IDF: 80% precision, 79% recall and f1-score with 0.79 which are not good as well

```python
###Try logistic regression model as classifier
from sklearn.linear_model import LogisticRegression
LR_clf = LogisticRegression(random_state=101)
LR_clf.fit(X_train, y_train)
predictions=LR_clf.predict(X_test)
print(classification_report(y_test, predictions))
```

precision of logistic regresion is 70%, and f1-score is 0.69, even worse than random forest
